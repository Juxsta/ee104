```python
import numpy as np
import scipy.integrate as integ
import matplotlib.pyplot as plt

```

$$ L\frac{\partial^{2}i}{\partial^{2}t}+R\frac{\partial i}{\partial t}+\frac{1}{C}i=0 $$

```python
def rlc(y,t,r,l,c):
    i,di=y
    return [di,-di*r/l-1/l/c*i]
```

```python
t = np.linspace(0,10,1000)
r,l,c=.1,.1,.1
y0 = [5,1]
sol = integ.odeint(rlc,y0,t,(r,l,c))
plt.plot(t,sol[:,0],label="i")

```
